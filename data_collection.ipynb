# DataSphere Scraper
Downloads all xlsx files from https://datasphere.uni-pr.edu/data/ and converts them to CSV.
Uses Playwright so JavaScript-rendered links and session cookies are handled automatically.

```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin
from pathlib import Path
import pandas as pd
import time, random

BASE_URL = "https://datasphere.uni-pr.edu/data/"
RAW_DIR  = Path("raw_data")
CSV_DIR  = Path("csv_data")
RAW_DIR.mkdir(exist_ok=True)
CSV_DIR.mkdir(exist_ok=True)
```

```python
# DL Balkans data

headers = {
    "User-Agent": (
        "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
        "AppleWebKit/537.36 (KHTML, like Gecko) "
        "Chrome/126.0.0.0 Safari/537.36"
    ),
    "Accept": (
        "text/html,application/xhtml+xml,application/xml;"
        "q=0.9,image/avif,image/webp,*/*;q=0.8"
    ),
    "Accept-Language": "en-US,en;q=0.9",
    "Referer": "https://datasphere.uni-pr.edu/",
    "Connection": "keep-alive",
}

session = requests.Session()
session.headers.update(headers)

response = session.get(BASE_URL, timeout=30)
response.raise_for_status()

soup = BeautifulSoup(response.text, "html.parser")

urls = []

for tag in soup.find_all("a", href=True):
    href = tag["href"]
    full_url = urljoin(BASE_URL, href)
    urls.append(full_url)

# Remove duplicates while preserving order
urls = list(dict.fromkeys(urls))


excel_urls = [
    url for url in urls
    if url.lower().endswith((".xlsx", ".xls", ".csv"))
]

import requests
from pathlib import Path
from urllib.parse import urlparse, unquote


download_dir = Path("raw_data/balkans")
download_dir.mkdir(exist_ok=True)

for url in excel_urls:
    filename = unquote(Path(urlparse(url).path).name)
    output_path = download_dir / filename

    print(f"Downloading {filename}...")

    response = requests.get(url, stream=True)
    response.raise_for_status()

    with open(output_path, "wb") as file:
        for chunk in response.iter_content(chunk_size=8192):
            if chunk:
                file.write(chunk)

    print(f"Saved to {output_path}")

print("All downloads complete.")
```

```python
# ── Step 2: convert each xlsx to csv (one csv per sheet) ─────────────────────

import os


def xlsx_to_csv(xlsx_path: Path, csv_dir: Path):
    xl = pd.ExcelFile(xlsx_path)
    sheets = xl.sheet_names
    print(f"  {xlsx_path.name}  ({len(sheets)} sheet(s))")
    for sheet in sheets:
        df = xl.parse(sheet)
        safe_sheet = sheet.replace("/", "_").replace("\\", "_").strip()
        csv_name = xlsx_path.stem + ".csv" if len(sheets) == 1 else f"{xlsx_path.stem}__{safe_sheet}.csv"
        out = csv_dir / csv_name
        df.to_csv(out, index=False)
        print(f"    -> {csv_name}  ({len(df):,} rows x {len(df.columns)} cols)")
os.makedirs(CSV_DIR / "kosovo", exist_ok=True)
os.makedirs(CSV_DIR / "balkans", exist_ok=True)

for xlsx_path in os.listdir(RAW_DIR / "kosovo"):
    xlsx_to_csv(Path(RAW_DIR) / "kosovo" / xlsx_path, CSV_DIR / "kosovo")

csv_files = list((CSV_DIR / "kosovo").glob("*.csv"))
print(f"\nTotal CSV files written: {len(csv_files)}")

for xlsx_path in os.listdir(RAW_DIR / "balkans"):
    xlsx_to_csv(Path(RAW_DIR) / "balkans" / xlsx_path, CSV_DIR / "balkans")

csv_files = list((CSV_DIR / "balkans").glob("*.csv"))
print(f"\nTotal CSV files written: {len(csv_files)}")
```
